```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

### Read in NIS data for both Laser & Radar

```python
df = pd.read_csv('NIS.csv', header=None, names=['Sensor', 'NIS'])
laser_data = df.where(df['Sensor']=='L').dropna()
radar_data = df.where(df['Sensor']=='R').dropna()
```

```python
laser_data.head()
```

```python
radar_data.head()
```

### Plot Data Using ùúí <sup>2</sup> .050

The values used in the plots are taken from the lecture video on the subject

```python
# Laser measures in 2 DOF => chi**2(0.050) = 5.991

fig, ax = plt.subplots(1,1, figsize=(10,10))
ax.plot(laser_data['NIS'], color='purple')
ax.axhline(5.991, color='orange')
```

```python
# Radar measures in 2 DOF => chi**2(0.050) = 7.815

fig, ax = plt.subplots(1,1, figsize=(10,10))
ax.plot(radar_data['NIS'], color='green')
ax.axhline(7.815, color='red')
```
