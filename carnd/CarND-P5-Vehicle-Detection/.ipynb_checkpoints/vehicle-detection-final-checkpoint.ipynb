# Vehicle Detection Project
--------------------------------------------------------------------------------
The goals / steps of this project are the following:

* Perform a extraction on a labeled training set of images and train a Linear SVM classifier
* Features to explore are: 
    1.) Color Histograms
    2.) Spatial bins of pixel values
    3.) Histogram of Oriented Gradients (HOG)
* Implement a sliding-window technique and use trained classifier to search for vehicles in images
* Create a heat map of detections and reject outliers
* Estimate a bounding box for vehicles detected
* Run the processing pipeline on a video stream and follow detected vehicles
--------------------------------------------------------------------------------

```python
def plot_image(img_list, titles=None, axes_off=True, imsize=(10,10), bgr=None, cmap=None):
    '''
    Utility for examining results from project functions. Plots images
    Input: img_list = List of images
           gray_maps = List of boolans indicating which of img_list is grayscale
           titles = List of strings indicating titles of each image
           axes_off = Boolean flag indicating if plot axes should be hidden
           imsize = Int tuple of image size
           bgr = List of booleans indicating which of img_list is in bgr
    Output: ax = axis object from plot
    '''
    if bgr == None:
        bgr = [False for i in range(len(img_list))]
    if cmap == None:
        cmap = [None for i in range(len(img_list))]
    
    if len(img_list)==1:
        f, ax = plt.subplots(1, 1, figsize=imsize)
        if bgr[0]:
            ax.imshow(img_list[0][:,:,::-1])
        if cmap[0] != None:
            ax.imshow(img_list[0], cmap=cmap[0])
        if (not bgr[0]) & (cmap[0] == None):
            ax.imshow(img_list[0])
        if axes_off:
            ax.axis('off')
        if titles != None:
            ax.set_title(titles[0], fontsize=20)
    else:
        f, ax = plt.subplots(1, len(img_list), figsize=imsize)
        f.tight_layout
        ax = ax.ravel()
        for i in range(len(img_list)):
            if bgr[i]:
                ax[i].imshow(img_list[i][:,:,::-1])
            if cmap[i] != None:
                ax[i].imshow(img_list[i], cmap=cmap[i])
            if (not bgr[i]) & (cmap[i] == None):
                ax[i].imshow(img_list[i])
            if axes_off:
                ax[i].axis('off')
            if titles != None:
                ax[i].set_title(titles[i], fontsize=20)
    return ax
```

### HOG Features

```python
def get_hog_features(img, orient, pix_per_cell, cell_per_block, 
                        vis=False, feature_vec=True):
    '''
    Shorthand for producing a histogram-oriented gradients (HOG) for an image
    Inputs: img = Image
            orient = Number of gradient orientations to bin values into
            pix_per_cell = Number of pixels per cell
            cell_per_block = Number of cells per block
            vis = Boolean to toggle return of an image of the graph
            feature_vec = Boolean to indicate whether this computation is meant to return a single feature vector
    Output: features (if vis) = HOG values
            hog_image (if vis) = Image of HOG
            output (if not vis) = HOG values
    '''
    
    # Call with two outputs if vis==True
    if vis == True:
        features, hog_image = hog(img, orientations=orient, 
                                  pixels_per_cell=(pix_per_cell, pix_per_cell),
                                  block_norm= 'L2-Hys',
                                  cells_per_block=(cell_per_block, cell_per_block), 
                                  transform_sqrt=True, 
                                  visualise=vis, feature_vector=feature_vec)
        return features, hog_image
    
    # Otherwise call with one output
    else:      
        features = hog(img, orientations=orient, 
                       pixels_per_cell=(pix_per_cell, pix_per_cell),
                       cells_per_block=(cell_per_block, cell_per_block), 
                       block_norm= 'L2-Hys',
                       transform_sqrt=True, 
                       visualise=vis, feature_vector=feature_vec)
        return features
```

### Spatial Features

```python
def bin_spatial(img, size=(32, 32)):
    '''
    Bin the pixels of an image into those of a smaller image using cv2's default bilinear interpolation
    Inputs: img = Image
            size = 2-tuple of binned image dimensions
    Output: Flattened array of pixel values from binned image
    '''
    
    # Use cv2.resize().ravel() to create the feature vector
    features = cv2.resize(img, size).ravel()
    return features
```

### Color Features

```python
def color_hist(img, nbins=32, bins_range=(0, 256)):
    '''
    Create a histogram of values for each color channel in an image
    Inputs: img = Image
            nbins = Number of bins to sort values into
            bins_range = Range of values image pixels may have
    Output: Array of histogram values
    '''
    
    # Compute the histogram of the color channels separately
    channel1_hist = np.histogram(img[:,:,0], bins=nbins, range=bins_range)
    channel2_hist = np.histogram(img[:,:,1], bins=nbins, range=bins_range)
    channel3_hist = np.histogram(img[:,:,2], bins=nbins, range=bins_range)
    
    # Concatenate the histograms into a single feature vector
    hist_features = np.concatenate((channel1_hist[0], channel2_hist[0], channel3_hist[0]))

    return hist_features
```

### Feature Extraction (used to train classifier)

```python
def extract_features(imgs, color_space='RGB', spatial_size=(32, 32),
                        hist_bins=32, orient=9, 
                        pix_per_cell=8, cell_per_block=2, hog_channel=0,
                        spatial_feat=True, hist_feat=True, hog_feat=True):
    '''
    Extract features from an array of images. Used to extract training data features
    Inputs: imgs = List of image filepaths
            color_space = Color space in which features should be extracted
            spatial_size = 2-Tuple indicating bin counts in x and y respectively
            hist_bins = Number of bins into which color values should be binned
            hist_range = Range of values image pixels may have
            orient = Number of gradient orientations to bin values into
            pix_per_cell = Number of pixels per cell in the HOG
            cell_per_block = Number of cells per block in the HOG
            hog_channel = Image color-channel(s) to use for computing HOG (Pass string 'ALL' to use all channels)
            spatial_feat = Boolean indicating whether or not spatial features should be extracted
            hist_feat = Boolean indicating whether or not color features should be extracted
            hog_feat = Boolean indicating whether or not HOG should be computed
    '''
    
    # Create a list to append feature vectors to
    features = []
    
    # Iterate through the list of images
    for file in imgs:
        file_features = []
        
        # Read in image
        image = mpimg.imread(file)
        
        # Apply color conversion if other than 'RGB'
        if color_space != 'RGB':
            feature_image = convert_color(image, 'RGB2'+color_space)
        else:
            feature_image = np.copy(image)      

        # Extract spatial features
        if spatial_feat == True:
            spatial_features = bin_spatial(feature_image, size=spatial_size)
            file_features.append(spatial_features)
        
        # Extract color features
        if hist_feat == True:
            hist_features = color_hist(feature_image, nbins=hist_bins)
            file_features.append(hist_features)
        
        # Compute HOG
        if hog_feat == True:
            if hog_channel == 'ALL':
                hog_features = []
                for channel in range(feature_image.shape[2]):
                    hog_features.append(get_hog_features(feature_image[:,:,channel], 
                                        orient, pix_per_cell, cell_per_block, 
                                        vis=False, feature_vec=True))
                hog_features = np.ravel(hog_features)        
            else:
                hog_features = get_hog_features(feature_image[:,:,hog_channel], orient, 
                            pix_per_cell, cell_per_block, vis=False, feature_vec=True)
            file_features.append(hog_features)
        features.append(np.concatenate(file_features))
    
    return features
```

### Window Generator

```python
def slide_window(img, x_start_stop=[None, None], y_start_stop=[None, None], 
                    xy_window=(64, 64), xy_overlap=(0.5, 0.5)):
    '''
    Generate a list of of coordinate pairs each of which identifies a window
    within an image
    Inputs: img = Image to window
            x_start_stop = List of min/max x coordinates within which windows will be generated
            y_start_stop = List of min/max y coordinates within which windows will be generated
            xy_window = 2-tuple specifying window size in x and y
            xy_overlap = 2-tuple specifying portion of window to overlap with on each step through the image
    Outputs: List of 2-tuples of 2-tuples containing top-left and bottom-right window coordinates
    '''
    # If x and/or y start/stop positions not defined, set to image size
    if x_start_stop[0] == None:
        x_start_stop[0] = 0
    if x_start_stop[1] == None:
        x_start_stop[1] = img.shape[1]
    if y_start_stop[0] == None:
        y_start_stop[0] = 0
    if y_start_stop[1] == None:
        y_start_stop[1] = img.shape[0]
        
    # Compute the span of the region to be searched   
    xspan = x_start_stop[1] - x_start_stop[0]
    yspan = y_start_stop[1] - y_start_stop[0]
    
    # Compute the number of pixels per step in x/y
    nx_pix_per_step = np.int(xy_window[0]*(1 - xy_overlap[0]))
    ny_pix_per_step = np.int(xy_window[1]*(1 - xy_overlap[1]))
    
    # Compute the number of windows in x/y
    nx_buffer = np.int(xy_window[0]*(xy_overlap[0]))
    ny_buffer = np.int(xy_window[1]*(xy_overlap[1]))
    nx_windows = np.int((xspan-nx_buffer)/nx_pix_per_step) 
    ny_windows = np.int((yspan-ny_buffer)/ny_pix_per_step) 
    
    # Initialize a list to append window positions to
    window_list = []
    
    # Loop through finding x and y window positions
    for ys in range(ny_windows):
        for xs in range(nx_windows):
            
            # Calculate window position
            startx = xs*nx_pix_per_step + x_start_stop[0]
            endx = startx + xy_window[0]
            starty = ys*ny_pix_per_step + y_start_stop[0]
            endy = starty + xy_window[1]
            
            # Append window position to list
            window_list.append(((startx, starty), (endx, endy)))

    return window_list
```

### Draw Boxes Given Windows

```python
def draw_boxes(img, bboxes, color=(0, 0, 255), thick=6):
    '''
    Shorthand for drawing boxes over an image, given a list of coordinates
    Inputs: img = Image to draw on
            bboxes = List of box coordinates
            color = RGB color values for box borders
            thick = Thickness of lines used to draw boxes
    Outputs: imcopy = Copy of input image with boxes drawn overtop
    '''
    # Make a copy of the image
    imcopy = np.copy(img)
    # Iterate through the bounding boxes
    for bbox in bboxes:
        # Draw a rectangle given bbox coordinates
        cv2.rectangle(imcopy, bbox[0], bbox[1], color, thick)
    # Return the image copy with boxes drawn
    return imcopy
```

### Apply Window Generator and Classify

```python
def find_cars(img, ystart, ystop, scale, svc, X_scaler,
              orient, pix_per_cell, cell_per_block, spatial_size, hist_bins, cspace):
    '''
    Extract features from a single image and pass values into a classifier. Record where in the image
    a positive classification was detected.
    Inputs: img = Image to search
            ystart = Lower limit in y-axis for searching
            ystop = Upper limit in y-axis for searching
            scale = Scale factor to apply to image while searching for features
            svc = Classifier
            X_scaler = Scaler used in training the classifier
            orient = Number of gradient orientations to bin values into
            pix_per_cell = Number of pixels per cell in the HOG
            cell_per_block = Number of cells per block in the HOG
            spatial_size = 2-Tuple indicating bin counts in x and y respectively
            hist_bins = Number of bins into which color values should be binned
            cspace = Color space in which the operations should be performed
    Outputs: draw_img = Copy of image with positively classified regions outlined
             bboxes = List of coordinates defining windows drawn on draw_img
    '''
    
    draw_img = np.copy(img)
    
    img_tosearch = img[ystart:ystop,:,:]
    ctrans_tosearch = convert_color(img_tosearch, conv='RGB2'+cspace)
    if scale != 1:
        imshape = ctrans_tosearch.shape
        ctrans_tosearch = cv2.resize(ctrans_tosearch, (np.int(imshape[1]/scale), np.int(imshape[0]/scale)))
        
    ch1 = ctrans_tosearch[:,:,0]
    ch2 = ctrans_tosearch[:,:,1]
    ch3 = ctrans_tosearch[:,:,2]

    # Define blocks and steps as above
    nxblocks = (ch1.shape[1] // pix_per_cell) - cell_per_block + 1
    nyblocks = (ch1.shape[0] // pix_per_cell) - cell_per_block + 1 
    nfeat_per_block = orient*cell_per_block**2
    
    # 64 was the orginal sampling rate, with 8 cells and 8 pix per cell
    window = 64
    nblocks_per_window = (window // pix_per_cell) - cell_per_block + 1
    cells_per_step = 2  # Instead of overlap, define how many cells to step
    nxsteps = (nxblocks - nblocks_per_window) // cells_per_step + 1
    nysteps = (nyblocks - nblocks_per_window) // cells_per_step + 1
    
    # Compute individual channel HOG features for the entire image
    hog1 = get_hog_features(ch1, orient, pix_per_cell, cell_per_block, feature_vec=False)
    hog2 = get_hog_features(ch2, orient, pix_per_cell, cell_per_block, feature_vec=False)
    hog3 = get_hog_features(ch3, orient, pix_per_cell, cell_per_block, feature_vec=False)
    
    bboxes = []
    
    for xb in range(nxsteps):
        for yb in range(nysteps):
            ypos = yb*cells_per_step
            xpos = xb*cells_per_step
            # Extract HOG for this patch
            hog_feat1 = hog1[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel() 
            hog_feat2 = hog2[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel() 
            hog_feat3 = hog3[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel() 
            hog_features = np.hstack((hog_feat1, hog_feat2, hog_feat3))

            xleft = xpos*pix_per_cell
            ytop = ypos*pix_per_cell

            # Extract the image patch
            subimg = cv2.resize(ctrans_tosearch[ytop:ytop+window, xleft:xleft+window], (64,64))
          
            # Get color features
            spatial_features = bin_spatial(subimg, size=spatial_size)
            hist_features = color_hist(subimg, nbins=hist_bins)

            # Scale features and make a prediction
            test_features = X_scaler.transform(np.hstack((spatial_features, hist_features, hog_features)).reshape(1, -1))    
            #test_features = X_scaler.transform(np.hstack((shape_feat, hist_feat)).reshape(1, -1))    
            test_prediction = svc.predict(test_features)
            
            if (test_prediction == 1) & (svc.decision_function(test_features) >= 0.75):
                xbox_left = np.int(xleft*scale)
                ytop_draw = np.int(ytop*scale)
                win_draw = np.int(window*scale)
                bbox = ((xbox_left, ytop_draw+ystart),(xbox_left+win_draw,ytop_draw+win_draw+ystart))
                draw_img = cv2.rectangle(draw_img,(xbox_left, ytop_draw+ystart),(xbox_left+win_draw,ytop_draw+win_draw+ystart),(0,0,255),6)
                bboxes.append(bbox)
                
    return draw_img, bboxes
```

### Produce Heatmap of Classifications

```python
def add_heat(heatmap, bbox_list):
    '''
    Add one to each heatmap pixel within a list of bounded areas
    Inputs: heatmap = Heatmap
            bboxes = List of 2-tuples of boxes' corner points (top-left and bottom-right)
    Outputs: Updated heatmap
    '''
    # Iterate through list of bboxes
    for box in bbox_list:
        # Add += 1 for all pixels inside each bbox
        # Assuming each "box" takes the form ((x1, y1), (x2, y2))
        heatmap[box[0][1]:box[1][1], box[0][0]:box[1][0]] += 1

    return heatmap
```

```python
test_image = mpimg.imread(glob.glob('images\\test*.jpg')[0])
searched_image, bboxes = find_cars(test_image, 400, 700, 1, svc, X_scaler,
              orient, pix_per_cell, cell_per_block, spatial_size, hist_bins, 'YUV')
heatmap = np.zeros_like(test_image[:,:,0])
heatmap = add_heat(heatmap, bboxes)
ax = plot_image([test_image, searched_image, heatmap],
                cmap=[None, None, 'hot'], titles=['Original', 'Search Results', 'Heatmap'], imsize=(20, 20))
```

### Threshold Heatmap

```python
def apply_threshold(heatmap, threshold):
    '''
    Low pass filter application to a heatmap
    Inputs: heatmap = Heatmap
            threshold = Minimum pixel value to pass the threshold
    '''
    # Zero out pixels below the threshold
    heatmap[heatmap <= threshold] = 0

    return heatmap
```

```python
test_image = mpimg.imread(glob.glob('images\\test*.jpg')[0])
searched_image, bboxes = find_cars(test_image, 400, 700, 1, svc, X_scaler,
              orient, pix_per_cell, cell_per_block, spatial_size, hist_bins, 'YUV')
heatmap = np.zeros_like(test_image[:,:,0])
heatmap = add_heat(heatmap, bboxes)
heatmap = apply_threshold(heatmap, 1)
ax = plot_image([test_image, searched_image, heatmap],
                cmap=[None, None, 'hot'], titles=['Original', 'Search Results', 'Thresholded Heatmap'], imsize=(20, 20))
```

### Cluster Heatmap

```python
def draw_labeled_bboxes(img, labels):
    '''
    Draw boxes around pixels corresponding to a set of previously identified labels
    Inputs: img = Image to draw on
            labels = List of labels and their corresponding pixels
    Outputs: Image with boxes drawn around labeled areas
    '''
    imcopy = np.copy(img)
    # Iterate through all detected cars
    for car_number in range(1, labels[1]+1):
        
        # Find pixels with each car_number label value
        nonzero = (labels[0] == car_number).nonzero()
        
        # Identify x and y values of those pixels
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        
        # Define a bounding box based on min/max x and y
        bbox = ((np.min(nonzerox), np.min(nonzeroy)), (np.max(nonzerox), np.max(nonzeroy)))
        
        # Draw the box on the image
        cv2.rectangle(imcopy, bbox[0], bbox[1], (0,0,255), 6)

    return imcopy
```

```python
test_image = mpimg.imread(glob.glob('images\\test*.jpg')[0])
searched_image, bboxes = find_cars(test_image, 400, 700, 1, svc, X_scaler,
              orient, pix_per_cell, cell_per_block, spatial_size, hist_bins, 'YUV')
heatmap = np.zeros_like(test_image[:,:,0])
heatmap = add_heat(heatmap, bboxes)
heatmap = apply_threshold(heatmap, 1)
labels = label(heatmap)
results = draw_labeled_bboxes(test_image, labels)
ax = plot_image([test_image, searched_image, heatmap],
                cmap=[None, None, 'hot'], titles=['Original', 'Search Results', 'Labeled Regions'], imsize=(20, 20))
```

## Final Pipeline

### Combine the preceding into single-image search function

```python
def search_image(img, ystart, svc, X_scaler,
                 orient, pix_per_cell, cell_per_block, spatial_size, hist_bins, cspace):
    '''
    Search through various regions of interest within a given image and apply find_cars
    Then, create a heatmap of positively classified regions
    Add the heatmap to a globally-defined deque and compute a running average of heatmaps
    Use this heatmap to identify where vehicles are in the current image
    Draw labels on the image and return
    Inputs: img = Image to search
            ystart = Lower limit in y-axis for searching
            svc = Classifier
            X_scaler = Scaler used in training the classifier
            orient = Number of gradient orientations to bin values into
            pix_per_cell = Number of pixels per cell in the HOG
            cell_per_block = Number of cells per block in the HOG
            spatial_size = 2-Tuple indicating bin counts in x and y respectively
            hist_bins = Number of bins into which color values should be binned
            cspace = Color space in which the operations should be performed
    Outputs: image = Copy of input image with boxes drawn around areas with many
                     positive classifications
    '''
    bboxes = []
    
    ystarts = [400, 400, 410, 420, 430] 
    ystops = [500, 500, 500, 556, 556]
    sfactors = [1, 1.3, 1.4, 1.6, 1.8]

    for ystart, ystop, scale in zip(ystarts, ystops, sfactors):

        im1, wins = find_cars(img, ystart, ystop, scale, svc, X_scaler,
                                  orient, pix_per_cell, cell_per_block, spatial_size, hist_bins, cspace)
        bboxes.extend(wins)
        
    ystarts = [430, 440, 400, 400, 500]
    ystops = [556, 556, 556, 556, 656]
    sfactors = [2, 1.9, 1.3, 2.2, 3]

    for ystart, ystop, scale in zip(ystarts, ystops, sfactors):

        im2, wins = find_cars(img, ystart, ystop, scale, svc, X_scaler,
                                orient, pix_per_cell, cell_per_block, spatial_size, hist_bins, cspace)
    
        bboxes.extend(wins)
    
    heat = np.zeros_like(img[:,:,0])
    
    # Add heat to each box in box list
    heat = add_heat(heat,bboxes)

    # Apply threshold to help remove false positives
    heat = apply_threshold(heat, 1)

    # Visualize the heatmap when displaying    
    heatmap = np.clip(heat, 0, 255)
    
    # Create a running average of heatmaps from last several images
    log.append(heatmap)
    heatmap_net = np.zeros_like(img[:,:,0])
    for record in range(len(log)-1):
        heatmap_net += log[record]  
    heatmap_net = heatmap_net//len(log)
    
    # Find final boxes from heatmap using label function
    labels = label(heatmap_net)
    
    image = draw_labeled_bboxes(img, labels)

    return image
```

### Condense into function for video frame analysis

```python
def video_process(image):
    '''
    Function to apply to image frames. Pass image and global parameters into search_image
    Inputs: image = Frame from video
    Outpus: Image returned from search_image with boxes drawn around vehicles
    '''
    image = search_image(image, ystart, svc, X_scaler,
                         orient, pix_per_cell, cell_per_block, spatial_size, hist_bins, colorspace)
    return image
```
